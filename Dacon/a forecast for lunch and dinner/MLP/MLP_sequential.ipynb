```python
# 텐서플로 ≥2.0 필수
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import *
assert tf.__version__ >= "2.0"

# 공통 모듈 임포트
import numpy as np
import os
import pandas as pd
```

```python
train_ln = pd.read_csv('/content/drive/MyDrive/코드리뷰/enco_train.csv')

test_ln = pd.read_csv('/content/drive/MyDrive/코드리뷰/enco_test.csv')
```

```python
train_ln.columns
```

```python
test_lunch = test_ln[['day','bob', 'soup', 'main','present']]
```

```python
test_dinner = test_ln[['day', 'bobd', 'soupd', 'maind','present']]
```

```python
# train_ln = [['day','bob','soup','main','present','lunch_t']]
# X_test = [['day','bob','soup','main','present']]
```

```python
X = train_ln[['day','bob', 'soup', 'main','present']]
y = train_ln['lunch_t']
```

```python

from sklearn.model_selection import train_test_split


X_train, X_vali, y_train, y_vali = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 

                                                    shuffle=True, 

                                                    random_state=1004)
```

```python
def model_dnn(Nin, Nh, Nout):
    model = keras.Sequential()
    model.add(keras.layers.Dense(Nh, activation='relu', input_shape=(Nin,)))
    for _ in range(100):
        model.add(keras.layers.Dense(Nh))
    model.add(keras.layers.Dense(Nh))
    model.add(keras.layers.Dense(Nout))
    model.compile(loss='mae',
                  optimizer='adam')
    return model
```

```python
model = model_dnn(X_train.shape[1],50,1)
```

```python
model.summary()
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=20)

history = model.fit(X_train, y_train,
                    validation_data=(X_vali, y_vali),
                    epochs=1000, verbose=1, batch_size=32, callbacks = [early_stopping_cb] )
```

```python
# 예측

lunch = model.predict(test_lunch)
```

```python
submit = pd.read_csv('/content/drive/MyDrive/코드리뷰/sample_submission.csv')
submit['중식계'] = lunch
```

```python
# 저녁

X_train, X_vali, y_train, y_vali = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 

                                                    shuffle=True, 

                                                    random_state=1004)

early_stopping_cb = keras.callbacks.EarlyStopping(patience=20)

history = model.fit(X_train, y_train,
                    validation_data=(X_vali, y_vali),
                    epochs=1000, verbose=1, batch_size=32, callbacks = [early_stopping_cb] )


dinner = model.predict(test_dinner)
```
